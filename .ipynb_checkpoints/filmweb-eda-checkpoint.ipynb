## Collecting data

Let's start from the beginning. We will crawl through https://filmweb.pl/ – the largest movie database right after IMDb – and collect data about movies and series, which we will later analyze and (hopefully) gain some insights.

In the first step, we will use scrapy library to create a simple crawler that will collect the necessary data and save it in JSON format.

If you want, you can get the data yourself, or simply use the sample used in this project (you will find it in the root directory).

```python
import scrapy
from scrapy.spiders import CrawlSpider, Rule
from scrapy.linkextractors import LinkExtractor

class FilmwebSpider(CrawlSpider):
    name = 'filmweb'
    allowed_domains = ['filmweb.pl']
    start_urls = ['https://www.filmweb.pl/']

    rules = (
        Rule(LinkExtractor(allow=('/film', '/serial')), 
             callback='parse', follow=True),
        )

    def parse(self, response):
        return {
            'url': response.url,
            'title': response.css('h1::text').get(),
            'org_title': response.css('div.filmCoverSection__originalTitle::text').get(),
            'year': response.css('div.filmCoverSection__year::text').get(),
            'duration': response.css('div.filmCoverSection__duration::attr(data-duration)').get(),
            'critic_rating': response.css('[data-source=criticRatingData]::text').get(),
            'ratings_dist': response.css('script[data-source=filmRating]::text').get(),
            'genre': response.css('div[itemprop=genre] > span > a::text').getall(),
            'country': response.xpath("//h3[contains(., 'produkcja')]/following-sibling::div/span/a/text()").getall(),
            'director': response.css('div[data-type=directing-info] > a::attr(title)').getall(),
            'screenwriter': response.css('div[data-type=screenwriting-info] > a::attr(title)').getall()
            }
```

## Data cleaning and wrangling

We already have everything we need. Unfortunately, the collected data is not ready for analysis. It needs to be cleaned first. We'll start with importing basic libraries and data itself.

```python
import pandas as pd
import numpy as np
import ast

raw = pd.read_json('filmweb_data.json', encoding='utf-8')
```

Let's glance over our data. We'll use .head() and .info() to see what we're dealing with.

```python
raw.head()
```

```python
raw.info()
```

Brief glance over show us our data set need few amendments:
- rows lacking movie data need to be dropped,
- columns 'critic_data' and 'ratings_dist' contain JSON data from original API, which require further extraction,
- movie duration needs to be converted from 'h godz. mm min.' to minutes,
- 'year' column contains mixed formats, single date for movies and two for series,
- data types need to be corrected.

Let's start with tagging movies and series based on URLs, and then we'll drop useless data.

```python
raw['type'] = ['Film' if '/film' in i else "Series" for i in raw['url']]

raw['type'].value_counts()
```

```python
raw.dropna(subset=['title', 'year', 'ratings_dist', 'type'], inplace=True)
raw.drop_duplicates(subset=['title', 'org_title', 'year', 'duration'], inplace=True)

raw.shape
```

Cool. There were 3825 rows without actual movie data. Now we can focus on columns containing JSONs with critics' ratings and votes distrubutions.

```python
tmp = raw.dropna(subset='critic_data')
tmp = tmp['critic_data'].apply(lambda x: ast.literal_eval(x))
tmp = tmp.apply(pd.Series)

raw['critic_rating'] = tmp['rate']
raw['critic_votes'] = tmp['count']

raw.head()
```

```python
tmp = raw.dropna(subset='ratings_dist')
tmp = tmp['ratings_dist'].str.strip(' ').apply(lambda x: ast.literal_eval(x))
tmp = tmp.apply(pd.Series)

raw['want_to_see'] = tmp['countWantToSee']
tmp.drop(['count', 'rate', 'countWantToSee'], axis=1, inplace=True)

raw = raw.join(tmp)

raw.columns
```

Now we need to extract the duration data, which is currently a string. To do this, we need to tear the string apart, extract the number of hours and minutes, and then convert them to minutes.

```python
tmp = pd.Series.to_frame(raw['duration'].str.split(' godz.'))
tmp.dropna(inplace=True)
tmp['hours'] = tmp['duration'].apply(lambda x: x.pop(0) if len(x) == 2 else 0)
tmp['minutes'] = pd.to_numeric(tmp['duration'].apply(lambda x: x.pop(0).strip(' min.')))
tmp['minutes'].fillna(0, inplace=True)

tmp['duration'] = tmp['hours'].astype('int') * 60 + tmp['minutes'].astype('int')
raw['duration_min'] = tmp['duration']

raw.head()
```

We'll do the same for the release year.

```python
tmp = pd.Series.to_frame(raw['year'].str.split(' - '))
tmp['end_date'] = tmp['year'].apply(lambda x: x.pop(1) if len(x) == 2 else None)
tmp['release_date'] = tmp['year'].apply(lambda x: x.pop(0))

raw['release_date'] = tmp['release_date']
raw['end_date'] = tmp['end_date']
raw
```

Before we'll build the final data frame and start exploration, let's straighten those data types. Firstly we'll look them up using .dtypes method. Next, we'll create a dictionary with column names paired with adequate data types.

```python
raw.dtypes
```

```python
data_types = {'community_votes':'Int64',
              'critic_votes':'Int64',
              'want_to_see':'Int64',
              'countVote1':'Int64',
              'countVote2':'Int64',
              'countVote3':'Int64',
              'countVote4':'Int64',
              'countVote5':'Int64',
              'countVote6':'Int64',
              'countVote7':'Int64',
              'countVote8':'Int64',
              'countVote9':'Int64',
              'countVote10':'Int64',
              'duration_min':'Int64',
              'release_date':'Int64',
              'end_date':'Int64'}

raw = raw.astype(data_types)
raw.dtypes
```

At last! We are ready to build the final dataframe. We have extracted all the data we needed and adjusted the data types. Now let's drop the data that's no longer needed and get down to the proper analysis.

```python
df = raw[['url', 'title', 'org_title', 'release_date', 'end_date', \
          'duration_min', 'genre', 'country', 'type', 'director', 'screenwriter', \
          'community_rating', 'community_votes', 'critic_rating', 'critic_votes', \
          'want_to_see', 'countVote1', 'countVote2', 'countVote3', 'countVote4', \
          'countVote5', 'countVote6', 'countVote7', 'countVote8', 'countVote9', 'countVote10']]

df.head()
```

## Data visualization

We already have the data, now we can start visualizing it. We will look at data and see if we can find something interesting along the way. In the first step, we will import the necessary libraries and establish a common figure size.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(15, 9)})
```

Let's start with the distribution of community ratings and compare it to the ratings of critics.

```python
sns.histplot(df['community_rating'], alpha=.5, bins=100, label='Community')
sns.histplot(df['critic_rating'], alpha=.5, bins=100, label='Critics', color='seagreen')

plt.title('Average movie rating')
plt.xlabel('Rating')
plt.xticks(range(1, 11))
plt.legend()
plt.show()
```

Interesting. We see left-skewed normal distribution here, but critics' voices have specific spikes. Since critics are a relatively small group among all users, we see a lot of movies with one or two ratings from critics, which translates into a large share of whole ratings. In addition, at first glance we can say that the critics are more strict in their judgments - the distribution is visibly shifted to the left.

Let's look at the ratings from different perspective.

```python
sns.catplot(data=pd.melt(df[['community_rating', 'critic_rating']]), 
            x='variable', y='value', 
            kind='box', whis=[5, 95], height=9)

plt.title('Average movie rating by community and critics')
plt.xlabel('')
plt.ylabel('Rating')
plt.yticks(range(1, 11))
plt.show()
```

As expected, critics rate movies more severely than rest of the community. The median of critics' ratings is around 6, while most community ratings are around 7. Critic ratings are also more varied, half of the ratings are between 5-7, while half of the community ratings are in the 6.1-7.4 range.

Now let's see how many movies were produced each year.

```python
sns.histplot(df['release_date'], bins=len(df['release_date'].unique()))

plt.title('Number of movies released each year')
plt.xlabel('Release year')
plt.xticks(range(df['release_date'].min(), 
                 df['release_date'].max() + 1, 10))
plt.show()
```

The number of movies is growing exponentially. Growth slowed down around 2000, peaking in 2014, and the number of releases has been declining since then.

Let's look at the distribution of movie durations, and then we'll count how many movies and series we have in the sample.

```python
sns.histplot(df[df['type'] == 'Film']['duration_min'], bins=100)

plt.title('Movie duration')
plt.xlabel('Duration in minutes')
plt.show()
```

```python
sns.countplot(data=df, x='type')

plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

Cool. It look like the majority of movies have running time around 100 minutes. 

Ok, let's see what genres and countries are best represented. To do this, we will break down the list of genres and countries into individual records and count how many movies are in each category.

```python
df_exp_genre = df.explode('genre').reset_index()

sns.barplot(x=df_exp_genre['genre'].value_counts().head(20).values, 
            y=df_exp_genre['genre'].value_counts().head(20).index)

plt.title('Number of movies per genre')

plt.show()
```

```python
df_exp_country = df.explode('country').reset_index()

sns.barplot(x=df_exp_country['country'].value_counts().head(20).values, 
            y=df_exp_country['country'].value_counts().head(20).index)

plt.title('Number of movies per country')
plt.show()
```

Not surprisingly, the most popular genres are drama and comedy, while the US is second to none when it comes to the number of movies produced.

We'll do the same with directors to see which of them are the most prolific.

```python
df_exp_director = df.explode('director').reset_index()

sns.barplot(x=df_exp_director['director'].value_counts().head(20).values, 
            y=df_exp_director['director'].value_counts().head(20).index)

plt.title('Number of movies per director')
plt.show()
```

Laurent Bouzerau is a real workhorse, 70 movies is an amazing score!

Now we can look at the distribution of ratings between genres and the most popular directors.

```python
most_watched_cats = list(df_exp_genre
                         .groupby('genre')['community_votes']
                         .sum()
                         .sort_values(ascending=False)
                         .head(10)
                         .index)

sns.catplot(data=df_exp_genre[(df_exp_genre['genre'].isin(most_watched_cats)) & 
                              (df_exp_genre['community_votes'] > 100)], 
            x='community_rating', y='genre', 
            order=most_watched_cats, 
            kind='box', whis=[5, 95], height=9)

plt.title('Average rating by genre')
plt.xlabel('Rating')
plt.xticks(range(1, 11))
plt.ylabel('')
plt.show()
```

```python
most_watched_dirs = list(df_exp_director.groupby('director')['community_votes']
                         .sum()
                         .sort_values(ascending=False)
                         .head(20)
                         .index)

sns.catplot(data=df_exp_director[(df_exp_director['director'].isin(most_watched_dirs)) & 
                                 (df_exp_director['community_votes'] > 100)], 
            x='community_rating', y='director', 
            order=most_watched_dirs, 
            kind='box', whis=[5, 95], height=9)

plt.title('Average movie rating by most popular directors')
plt.xlabel('Rating')
plt.ylabel('')
plt.show()
```

All genres have fairly even ratings, except for horror. Scary movies get significantly lower rates - interesting, is it fear or cringe? 

Frank Darabont stands out among the most popular directors - it shouldn't come as a surprise to anyone who watched "The Green Mile" and "The Shawshank Redemption".


For as long as I can remember, old movies have been a benchmark of quality. Despite smaller budgets, poorer image quality, lack of CGI and other benefits of technological development – you can still find comments about "good ol' times" under new movie trailers.

Let's see if we can find this in the data.

```python
over_100 = df['community_votes'] > 100

sns.lineplot(x=df[over_100]['release_date'], 
             y=df[over_100]['community_rating'], 
             ci=None, label='Community')

sns.lineplot(x=df[over_100]['release_date'], 
             y=df[over_100]['critic_rating'], 
             ci=None, label='Critics')

plt.title('Average ratng by release date')
plt.xticks(range(df[over_100]['release_date'].min(), 
                 df[over_100]['release_date'].max() + 1, 10))
plt.yticks(range(1, 11))
plt.ylabel('Rating')
plt.xlabel('Release date')
plt.legend()

plt.show()
```

We have it! Around 1960, the average ratings of films began to decline, with critics again being harsh judges - their ratings dropped more sharply than other users.

Let's see if we see any correlation between the average rating and the duration of the movie.

```python
sns.jointplot(data=df[df['type'] == 'Film'], 
              x='duration_min', y='community_rating', 
              kind='hex', height=12)

plt.title('Average rating vs movie duration')
plt.xlabel('Duration in minutes')
plt.ylabel('Community rating')
plt.yticks(range(1, 11))
plt.show()
```

There seems to be a slight correlation. It seems that longer videos tend to have a slightly higher rating. Bearing in mind the earlier plots of the average rating by genre, I'd like to break down the graph into a few selected genre layers.

```python
sns.jointplot(data=df_exp_genre[(df_exp_genre['type'] == 'Film') & 
                                (df_exp_genre['genre'].isin(['Dramat', 'Komedia', 'Horror'])) &
                                (df_exp_genre['community_votes'] > 100)], 
              x='duration_min', y='community_rating', 
              hue='genre', s=50, alpha=.5, height=12)

plt.yticks(range(1, 11))
plt.show()
```

As you can see, this slight correlation was completely random. Dramas are generally longer and better rated movies than shorter comedies and horrors.

Finally, we'll see if there are any differences in the distribution of votes in movies that critics rate better than the rest of community.

```python
df_critics = df['critic_rating'] > df['community_rating']
df_votes_critics = df[df_critics].loc[:, 'countVote1':'countVote10']
df_votes_community = df[~df_critics].loc[:, 'countVote1':'countVote10']

df_votes_critics_n = df_votes_critics.sum() / df_votes_critics.sum().sum()
df_votes_community_n = df_votes_community.sum() / df_votes_community.sum().sum()

sns.barplot(x=df_votes_community_n.index, y=df_votes_community_n.values, color='skyblue', alpha=.5, label='Community')
sns.barplot(x=df_votes_critics_n.index, y=df_votes_critics_n.values, color='lightcoral', alpha=.5, label='Critics')

plt.title('Votes distribution by movies higher rated by community and critics')
plt.xlabel('')
plt.ylabel('Fraction')
plt.legend()
plt.show()
```

Interesting. We can see that the distribution of ratings in the movies rated higher by critics is shifted to the right - these films get more 8s, 9s and 10s. Do the critics really have a nose for good films, or are we seeing users inspired by their ratings?
